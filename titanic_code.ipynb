```python
import pandas as pd 
import numpy as np 
```

```python
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('~/Desktop/titanic/train.csv')
df.drop(['Name', 'PassengerId', 'Ticket', 'Cabin'], axis=1, inplace=True)

train= pd.read_csv('~/Desktop/titanic/test.csv')
X_test = train.drop(['Name', 'PassengerId', 'Ticket','Cabin'], axis=1)

lb_enc = LabelEncoder()

def encode_with_unseen_handling(column, encoder):
    return column.apply(lambda x: encoder.transform([x])[0] if x in encoder.classes_ else -1)

# transform sex using label encoding
lb_enc.fit(df['Sex'])
df['Sex'] = lb_enc.transform(df['Sex'] )
X_test['Sex'] = lb_enc.transform(X_test['Sex'])

# transform embarked using label encoding
lb_enc.fit(df['Embarked'])
df['Embarked'] = lb_enc.transform(df['Embarked'])
X_test['Embarked'] = lb_enc.transform(X_test['Embarked'])



```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Survived', axis=1)
y = df['Survived']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.3, random_state=47)
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier

rf = HistGradientBoostingClassifier(learning_rate=0.1, random_state=43, l2_regularization=0.50)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
```

```python
submission = pd.DataFrame({'PassengerId':train['PassengerId'], 'Survived':rf_pred})
submission.to_csv('~/Desktop/titanic/submission.csv', index=False)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score

rf_pred = rf.predict(X_val)
roc_auc_score(y_val, rf_pred)
```

```python
f1_score(y_val, rf_pred)
```
