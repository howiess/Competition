## Motion Deception Metric (Motion-X)





#### Defining the goal
**Goal**: Quantify the effectiveness of pre-snap motion in terms of its ability to:
+ Distrupt the defensive alignment (deception)
+ Create offensive advantages (e.g. improved play success metrics)

**Metrics to measure**: Defensive displacement, reaction delays, play success metrics (e.g., yards gained, EPA, completion percentage)

**Refine the Objective**
+ Quantify the impact of pre-snap motion on defensive reactions and offensive success using available NFL tracking and play data.

**Key Questions**:
+ How does pre-snap motion influence defender alignment and coverage?
+ Does pre-snap motion result in a measurable delay in defender reaction time?
+ How does the presence of motion correlate with play success metrics (e.g., EPA, yards gained)?

**Data Source**: 
+ tracking data
+ play data 
+ player-play data
+ player data 
+ game data


**Key Variables**:
+ Pre-Snap Motion:
    + Use inMotionAtBallSnap or motionSinceLineset to identify motion events
    + Use tracking data to calculate the displacement of players flagged as "in motion"
+ Defender Reaction and Realignment:
    + Measure lateral movement (x, y) or acceleration in defenders relative to their starting position
    + Look for defensive coverage shifts (e.g., pff_passCoverage, pff_manZone, and pff_defensiveCoverageAssignment)
+ Play Success:
    + Metrics: expectedPointsAdded, yardsGained, completion percentage (passResult == C)
    + Normalize outcomes by play context (e.g., down, distance, field position)
+ Control Variables:
    + Formation: Offense (offenseFormation) and defense (pff_passCoverage)
    + Play Type: Pass, run, RPO, etc. (isDropback, pff_runPassOption)

#### **Step 1: Data Preparation**

**Join Datasets**:
+ Merge tracking_week_[week].csv, plays.csv, and player_play.csv using gameId, playId, and nflId
+ Enrich with players.csv for player attributes and games.csv for game context

**Filter for Motion Plays**:
+ Select plays where motionSinceLineset == True or inMotionAtBallSnap == True
+ Exclude penalties (playNullifiedByPenalty == Y) and special teams


#### **Step 2: Identify Defensive Reaction Metrics**

**Defender Displacement**:
+ Use tracking data to calculate the difference in defender positions (x, y) between:
    + Lineset Position: Position when the offense is initially set
    + Snap Position: Position at ball snap
+ Compute aggregate displacement across defenders

**Reaction Time**:
+ Measure time from the ball snap (event == snap) to:
    + Defender acceleration above a threshold (e.g., a > 0.5 y/s²)
    + Defender lateral movement crossing a threshold (e.g., x changes by 1+ yards)

**Coverage Shift**:
+ Compare pre-motion vs. post-motion coverage (pff_passCoverage, pff_defensiveCoverageAssignment)


#### **Step 3: Correlate Motion with Play Success**

**Create Motion-X Score**:
+ Formula: Motion-X = (Defender Displacement × Reaction Delay × Play Success Multiplier)
+ Normalize each component:
    + Displacement: Average yards moved
    + Reaction Delay: Time in seconds
    + Play Success Multiplier: Outcome-based weighting (e.g., yards gained, EPA, completion)

**Regression Analysis**:
+ Dependent Variable: Play success metrics (expectedPointsAdded, yardsGained)
+ Independent Variables:
    + Motion-X score
    + Control variables: offenseFormation, down, yardsToGo, pff_passCoverage


#### **Step 4: Build Predictive Models**

**Machine Learning Models**:
+ Train a supervised model (e.g., Random Forest, XGBoost) to predict play success from:
    + Defensive reaction metrics
    + Play context (down, distance)
    + Motion characteristics (speed, distance, direction)

**Feature Importance**:
+ Assess which aspects of motion (e.g., direction, speed) have the largest impact on play success


#### **Step 5: Visualize the Findings**

**Heatmaps**:
+ Show defender displacement before and after snap for motion plays.

**Flow Maps**:
+ Visualize player trajectories during motion, highlighting defensive reactions.

**Comparative Charts**:
+ Motion vs. non-motion plays for metrics like EPA, yards gained, and defender displacement.

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Circle

# Function to interpolate points
def interpolate_points(start, end, num_frames):
    """Interpolate between start and end points."""
    x_interp = np.linspace(start[0], end[0], num_frames)
    y_interp = np.linspace(start[1], end[1], num_frames)
    return x_interp, y_interp

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 5.33))
ax.set_xlim(0, 120)
ax.set_ylim(0, 53.3)
ax.set_aspect('equal')

# Set grid and labels
ax.set_title('Players Movement Animation')
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_xticks(np.arange(10, 120, 10))
ax.set_xticklabels(['G', 10, 20, 30, 40, 50, 40, 30, 20, 10, 'G'])
ax.grid(axis='x', linestyle='-')
ax.axvline(x=10, color='orange', linestyle='-', linewidth=2)
ax.axvline(x=110, color='orange', linestyle='-', linewidth=2)

# Prepare the legend
club_names = view['club'].unique()
teams = [club for club in club_names if club != "football"]
team_1, team_2 = teams[0], teams[1]

team_colors = {team_1: 'red', team_2: 'blue', 'football': 'brown'}

# Placeholder for patches
markers = []

# Animation initialization function
def init():
    for marker in markers:
        ax.add_patch(marker)
    return markers

# Animation update function
def update(frame):
    for i, marker in enumerate(markers):
        x_pos, y_pos = interpolated_frames[i][frame]
        marker.center = (x_pos, y_pos)
    return markers

# Prepare interpolated frames for all points
def generate_interpolated_frames(view, fps=10):
    """Generate interpolated frames for all points."""
    interpolated_frames = []

    for name, group in view.groupby("displayName"):
        x = group["x"].to_numpy()
        y = group["y"].to_numpy()

        # Interpolate each segment
        num_frames_per_segment = int(0.1 * fps)
        x_segments, y_segments = [], []

        for i in range(len(x) - 1):
            x_interp, y_interp = interpolate_points(
                (x[i], y[i]), (x[i + 1], y[i + 1]), num_frames_per_segment
            )
            x_segments.append(x_interp)
            y_segments.append(y_interp)

        # Concatenate all segments for this point
        x_interp_total = np.concatenate(x_segments)
        y_interp_total = np.concatenate(y_segments)

        interpolated_frames.append(list(zip(x_interp_total, y_interp_total)))

    return interpolated_frames

# Animation function
def animate_points(view, fps=10):
    global interpolated_frames, markers

    # Generate interpolated frames for all points
    interpolated_frames = generate_interpolated_frames(view, fps)

    # Initialize markers with appropriate types and sizes
    for name, group in view.groupby("displayName"):
        color = team_colors[group['club'].iloc[0]]
        x_pos, y_pos = interpolated_frames[0][0]  # Initial position
        
        if name == "football":
            # Smaller circle for football
            circle = Circle((x_pos, y_pos), radius=0.6, color=color)
        else:
            # Larger circle for players
            circle = Circle((x_pos, y_pos), radius=0.8, color=color)

        markers.append(circle)
        ax.add_patch(circle)

     # Add team-based legend
    team_legend = [
        plt.Line2D([0], [0], color=team_colors[team_1], label=team_1, markersize=10),
        plt.Line2D([0], [0], color=team_colors[team_2], label=team_2, markersize=10),
        plt.Line2D([0], [0], color=team_colors['football'], label='Football', markersize=10)
    ]
    ax.legend(handles=team_legend, loc="upper left")

    # Create animation
    total_frames = len(interpolated_frames[0])  # Assuming all points have the same length
    anim = FuncAnimation(
        fig, update,
        init_func=init,
        frames=total_frames,
        interval=1000 / fps,  # Interval in milliseconds
        blit=True,
        repeat=False
    )

    plt.tight_layout()
    plt.rcParams['animation.html'] = 'jshtml'
    return anim

# Animate
anim = animate_points(view)
anim
```
